```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics
import math 
```

```python
def net_visual(net, name):

  net_test = H_Test(net)

  x_axis = np.arange(-0.4e-01, 1e-01, 0.001)
  keys = [*net]

  

  plt.figure(figsize=(12,7))
  plt.plot(x_axis, norm.pdf(x_axis, net_test.mean()[keys[0]],
                            math.sqrt(net_test.var()[keys[0]])),
          label = keys[0][len(name)+1:])
  plt.plot(x_axis, norm.pdf(x_axis, net_test.mean()[keys[1]],
                            math.sqrt(net_test.var()[keys[1]])),
  label = keys[1][len(name)+1:])
  plt.plot(x_axis, norm.pdf(x_axis, net_test.mean()[keys[2]],
                            math.sqrt(net_test.var()[keys[2]])),
  label = keys[2][len(name)+1:])
  plt.plot(x_axis, norm.pdf(x_axis, net_test.mean()[keys[3]],
                            math.sqrt(net_test.var()[keys[3]])),
  label = keys[3][len(name)+1:])
  plt.plot(x_axis, norm.pdf(x_axis, net_test.mean()[keys[4]],
                            math.sqrt(net_test.var()[keys[4]])),
  label = keys[4][len(name)+1:])
  plt.plot(x_axis, norm.pdf(x_axis, net_test.mean()[keys[5]],
                            math.sqrt(net_test.var()[keys[5]])),
  label = keys[5][len(name)+1:])
  plt.plot(x_axis, norm.pdf(x_axis, net_test.mean()[keys[6]],
                            math.sqrt(net_test.var()[keys[6]])),
           label = keys[5][len(name)+1:])
  
  plt.title(name + ' trained with no noise and with the six noise types')
  plt.legend()


```

```python
def noise_visual(noise, t):
  noise_test = H_Test(noise)

  x_axis = np.arange(-0.4e-01, 1e-01, 0.001)
  keys = [*noise]
  
  
  plt.figure(figsize=(12,7))
  plt.plot(x_axis, norm.pdf(x_axis, noise_test.mean()[keys[0]],
                            math.sqrt(noise_test.var()[keys[0]])),
          label = keys[0][:-len(t)])
  plt.plot(x_axis, norm.pdf(x_axis, noise_test.mean()[keys[1]],
                            math.sqrt(noise_test.var()[keys[1]])),
            label = keys[1][:-len(t)])
  plt.plot(x_axis, norm.pdf(x_axis, noise_test.mean()[keys[2]],
                            math.sqrt(noise_test.var()[keys[2]])),
          label = keys[2][:-len(t)])
  
  plt.title('The three networks trained with '+t+' noise')

  plt.legend()

```

```python
for net in [[alexnet,'alexnet'],[resnet,'resnet'],[squeezenet,'squeezenet']]:
  net_visual(net[0],net[1])
```

```python
for noise in [[speckle, 'speckle'],[gaussian, 'gaussian'],
              [fgsm,'fgsm'],[s_and_p,'s&p'],[pgd,'pgd'],[eraser,'eraser']]:
        noise_visual(noise[0],noise[1])
```
